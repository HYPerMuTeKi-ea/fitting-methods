```python
import numpy as np
import matplotlib.pyplot as plt
c1=np.array([100.5,110,96.5,136,129.5,111.5,111.0])
v1=np.array([440,490,540,590,640,690,700])
c1_err=np.array([10,10.5,9.8,11.7,11.4,10.6,10.5])
v1_err=10
coefficient=np.polyfit(v1,c1,3)
polynomial=np.poly1d(coefficient)
xfit=np.linspace(440,700,100)
yfit=polynomial(xfit)
# 创建图表
plt.figure(figsize=(10, 6))

# 绘制带误差棒的原始数据
plt.errorbar(v1, c1, xerr=v1_err,yerr=c1_err, fmt='o', label='Data with error bars', capsize=5)

# 绘制线性回归拟合线
plt.plot(xfit, yfit, color='red')

# 添加图例和标签

plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# 显示图表
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 生成模拟数据
np.random.seed(0)
x = np.array([0,0.35,0.6,0.9,1.2,1.3,1.4])
y = np.array([400,450,500,550,600,650,700]) # 生成线性数据并添加噪声
y_err = 10  # 生成误差
x_err=0.1
# 使用scipy进行线性回归
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# 预测y值
y_pred = slope * x + intercept

# 创建图表
plt.figure(figsize=(10, 6))

# 绘制带误差棒的原始数据
plt.errorbar(x, y, xerr=x_err,yerr=y_err, fmt='o', label='Data with error bars', capsize=5)

# 绘制线性回归拟合线
plt.plot(x, y_pred, color='red', label=f'Linear fit: y={slope:.2f}x+{intercept:.2f}')

# 添加图例和标签
plt.title('Linear Regression with Error Bars')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
#打印slope 和intercept即可。
print(intercept,slope)
# 显示图表
plt.show()
```

```python
# 生成一些样本数据
np.random.seed(42)
X =np.array([3,4,2,5,1,6,7,0,8,10])

y =np.array([112,92,90,78,46,38,19,12,11,1])

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import poisson

# 你的数据
data = a

# 计算数据的直方图
bins = np.arange(0, max(data) + 1.5) - 0.5
hist, bin_edges = np.histogram(data, bins=bins, density=True)
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])

# 定义泊松分布的概率质量函数
def poisson_pmf(k, lamb):
    return poisson.pmf(k, lamb)

# 使用曲线拟合方法来估计λ（lambda）
params, cov_matrix = curve_fit(poisson_pmf, bin_centers, hist, p0=[np.mean(data)])

# 输出拟合的λ参数
lambda_est = params[0]
print(f"拟合出的 λ 参数: {lambda_est:.2f}")

# 绘制直方图和拟合曲线
plt.figure(figsize=(10, 6))
plt.hist(data, bins=bins, density=True, alpha=0.6, color='g', label='Data')

# 绘制拟合的泊松分布
x_plot = np.arange(0, max(data) + 1)
plt.plot(x_plot, poisson.pmf(x_plot, lambda_est), 'r-', lw=2, label=f'Poisson fit: λ={lambda_est:.2f}')

# 添加图例和标签
plt.title('Poisson Distribution Fit')
plt.xlabel('Number of events')
plt.ylabel('Probability')
plt.legend()

# 显示图表
plt.show()
```
